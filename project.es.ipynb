#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
def normalize_minmax(data):
    return (data - data.min()) / (data.max() - data.min())

# Normalizar precios para cada población
for population in cinturon_sur:
    ds_cinturon_sur.loc[ds_cinturon_sur['level5'] == population, 'price_normalized'] = normalize_minmax(ds_cinturon_sur.loc[ds_cinturon_sur['level5'] == population, 'price'])

fig, ax = plt.subplots()

for population in cinturon_sur:
    ax.hist(ds_cinturon_sur.loc[ds_cinturon_sur['level5'] == population, 'price_normalized'], alpha=0.5, label=population)

ax.set_xlabel('Precio normalizado')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución de precios normalizados en el cinturón sur')
ax.legend()

plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python

# Filtrar sin cero ni NaN
ds_cinturon_sur = ds_cinturon_sur[(ds_cinturon_sur['surface'] > 0) & (~ds_cinturon_sur['surface'].isna())]

# Crear pps
ds_cinturon_sur['pps'] = ds_cinturon_sur['price'] / ds_cinturon_sur['surface']

# Filtrar las ciudades de Getafe y Alcorcón
getafe_alcorcon = ds_cinturon_sur[ds_cinturon_sur['level5'].isin(['Getafe', 'Alcorcón'])]

# Verificar si hay datos para Getafe y Alcorcón
if not getafe_alcorcon.empty:
    # analizar de pps
    print(getafe_alcorcon.groupby('level5')['pps'].describe())
    
    # Extraer valores para boxplot
    getafe_pps = getafe_alcorcon.loc[getafe_alcorcon['level5'] == 'Getafe', 'pps'].values
    alcorcon_pps = getafe_alcorcon.loc[getafe_alcorcon['level5'] == 'Alcorcón', 'pps'].values
    
    # Verificar si hay datos suficientes para graficar
    if len(getafe_pps) == 0 or len(alcorcon_pps) == 0:
        print("No hay datos suficientes para graficar.")
    else:
        # Crear el boxplot
        fig, ax = plt.subplots()
        ax.boxplot([getafe_pps, alcorcon_pps], labels=['Getafe', 'Alcorcón'])
        plt.title('Precio por metro cuadrado en Getafe y Alcorcón')
        plt.xlabel('Municipio')
        plt.ylabel('Precio por metro cuadrado (€/m²)')
        plt.show(block=True)  # Forzar la visualización
else:
    print("No hay datos para Getafe o Alcorcón.")
    
    #
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa. No funciona, toma el subconjunto del ejercicio 14, crea un codigo simple que transforme columnas en coordenadas de diccionario de Python para agregarla a un mapa con columnas de latitud y longitud

```python
from ipyleaflet import Map, Marker, MarkerCluster


# 1. Filtrar el DataFrame (subconjunto del Ejercicio 14)
cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_cinturon_sur = ds[ds['level5'].isin(cinturon_sur)]

# 2. Crear un diccionario de coordenadas agrupadas por población
coordenadas_por_poblacion = ds_cinturon_sur.groupby('level5')[['latitude', 'longitude']].apply(lambda x: x.values.tolist()).to_dict()

```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
# 3. Crear el mapa centrado en Madrid
mapa = Map(center=(40.3, -3.7), zoom=11)

# 4. almacenar los marcadores
marcadores = []

# 5. Iterar sobre cada población y sus coordenadas
for poblacion, coordenadas in coordenadas_por_poblacion.items():
    # Crear un marcador para cada coordenada
    for coord in coordenadas:
        marcador = Marker(
            location=coord,  # Coordenadas (latitud, longitud)
            draggable=False  # No permitir arrastrar el marcador
        )
        marcadores.append(marcador)  # Agregar el marcador a la lista

# 6. Crear un grupo de marcadores (MarkerCluster)
marker_cluster = MarkerCluster(markers=marcadores)

# 7. Agregar el grupo de marcadores al mapa
mapa.add_layer(marker_cluster)

# 8. Mostrar el mapa
mapa
```
